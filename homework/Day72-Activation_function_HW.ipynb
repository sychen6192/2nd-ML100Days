# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline


def ReLU(x):
    return abs(x) * (x > 0)

def dReLU(x):
    return 1 * (x > 0)
```

```python
x = plt.linspace(-10, 10, 100)
plt.grid()
plt.title('ReLU vs dReLU')
plt.plot(x, ReLU(x), 'b', label='linspace(-10,10,10)')
plt.plot(x, dReLU(x), 'r')
plt.text(-10, 9.2,   r'ReLU(x) = abs(x) , if (x > 0)', fontsize=12)
plt.text(-10, 8.2,   r'dReLU(x) = 1 , if (x > 0)', fontsize=12)
#resize the X and Y axes (Locator(間隔))
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
```
